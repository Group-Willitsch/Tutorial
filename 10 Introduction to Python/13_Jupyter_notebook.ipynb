```python
## Library imports

import numpy as np
import matplotlib.pyplot as plt
```

# This is a markdown cell explaining the functions defined below

- `lin(x,a,b)` returns $ax + b$.
- `quad(x,a,b,c)` returns $ax^2 + bx + c$.
- `fluo_profile(x,a,b)` returns the scattering rate of a two-level system $$K_0\frac{\Gamma^2}{4} \frac{1}{\frac{\Gamma^2}{4} + (\delta - \mathbf{k}\cdot\mathbf{v}\cos{\omega t - \phi})^2}.$$


You see you can put text and $\LaTeX$ together and **format** your _text_ using markdown *formatting* !

```python
## Functions definitions

def lin(x,a,b):
    return a*x+b

def quad(x,a,b,c):
    return a*x**2 + b*x + c

def fluo_profile(t,v,K0,Gamma,delta,k,omega,phi):
    return K0*Gamma**2/4 / (Gamma**2/4 + (delta - k*v*np.cos(40*np.pi/180)*np.cos(omega*t-phi))**2)
```

```python
## Plot some of the functions
## defined above

x = np.linspace(-1,1,100)

plt.figure('linear')
plt.plot(x, lin(x,1,3))
plt.plot(x,quad(x,1,0,2))
plt.grid()
plt.tight_layout()
```

```python
## Plot fluo vs time
## for single trapped ion

# define physical variables

v = 5
K0 = 1000
Gamma = 21.57*1e6*2*np.pi
delta = -14*Gamma
wavelength = 397.959140*1e-9
k = 2*np.pi/wavelength
omega = 422.5*1e3*2*np.pi
phi = 0

t = np.linspace(0,3/omega*(2*np.pi),1000)*1e6

# plot fluo_profile

plt.figure('fluo_Ca')

plt.plot(t, fluo_profile(t,v,K0,Gamma,delta,k,omega,phi))

plt.xlabel('Time (us)')
plt.ylabel('Fluorescence (cts)')
plt.grid()
plt.tight_layout()
```
