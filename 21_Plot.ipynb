# 21_Plot

This program demonstrates how to plot data in different ways.
- Plot two functions directly in a figure.
- Plot two functions in a unique subplot in the figure.
- Plot two functions in two separated subfigures.
- Plot two functions in two separated subfigures in a separated window.

```python
## Library imports

import numpy as np
import matplotlib.pyplot as plt
```

```python
## Functions definitions

def lin(x,a,b):
    return a*x+b

def quad(x,a,b,c):
    return a*x**2 + b*x + c
```

```python
## Plot some of the functions
## defined above

# define x axis
x = np.linspace(-1,1,100)

# setting the plot
plt.figure('plot1')

# plot each function
plt.plot(x, lin(x,1,3),label='lin')
plt.plot(x,quad(x,1,0,2),label='quad')

# plot parameters
plt.xlabel('x')
plt.ylabel('y')

plt.grid()
plt.legend()
plt.tight_layout()
```

```python
## Plot some of the functions
## using subplots
## in same plot

# define x axis
x = np.linspace(-1,1,100)

# setting the plot
fig, ax = plt.subplots(num='plot2')

# plot each function
ax.plot(x, lin(x,1,3),label='lin')
ax.plot(x,quad(x,1,0,2),label='quad')

# plot parameters
# notice how the calls change a little bit
ax.set_xlabel('x')
ax.set_ylabel('y')

ax.grid()
plt.legend()
plt.tight_layout()
```

```python
help(plt.subplots)
```

```python
## Plot some of the functions
## using subplots
## in separate plots

# define x axis
x = np.linspace(-1,1,100)

# setting the plot
fig, ax = plt.subplots(1,2,num='plot2')

# plot each function
ax[0].plot(x, lin(x,1,3),label='lin')
ax[1].plot(x,quad(x,1,0,2),label='quad')

# plot parameters
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].grid()
ax[0].legend()

ax[1].set_xlabel('x')
ax[1].set_ylabel('y')
ax[1].grid()
ax[1].legend()

plt.tight_layout()
```

```python
## Plot some of the functions
## using subplots
## in separate plots

# define x axis
x = np.linspace(-1,1,100)

# setting the plot
# Notice again you have to change few things
# try to execute this cell again and again without clear=True
fig, ax = plt.subplots(1,2,num='plot2',clear=True)

# plot each function
ax[0].plot(x, lin(x,1,3),color='xkcd:fire engine red',label='lin')
ax[1].plot(x,quad(x,1,0,2),color='xkcd:sickly green',label='quad')

# plot parameters
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].grid()
ax[0].legend()

ax[1].set_xlabel('x')
ax[1].set_ylabel('y')
ax[1].grid()
ax[1].legend()

plt.tight_layout()## 
```
